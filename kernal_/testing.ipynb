```python
import re
from pdfminer.high_level import extract_pages
from pdfminer.layout import LTTextBox, LTTextLine
import pandas as pd

specific_x = 100.560
text_input = []
coordinate = []

for page_layout in extract_pages(pdf, page_numbers=[page]):
    for element in page_layout:
        if isinstance(element, (LTTextBox, LTTextLine)):
            if specific_x == element.x0:
                #text_input.extend(element.get_text().split("\n"))
                text_input.append(element.get_text())
                coordinate.append(element.y0)

df = pd.DataFrame({"Beskrivelse": text_input, "Coordinate": coordinate})
df.head()
```
